# Incremental Data Loading using AutoLoader

```python
%sql
CREATE SCHEMA IF NOT EXISTS netflix_catalog.net_shema;
```

```python
checkpoint_path = "abfss://silver@projectdatalakestrgacc.dfs.core.windows.net/checkpoint"
```

```python


df = spark.readStream\
  .format("cloudFiles")\
  .option("cloudFiles.format", "csv")\
  .option("cloudFiles.schemaLocation", checkpoint_path)\
  .load("abfss://raw@projectdatalakestrgacc.dfs.core.windows.net/")
```

```python
display(df)
```

```python
  
  df.writeStream\
  .option("checkpointLocation", checkpoint_path)\
  .trigger(processingTime='10 seconds')\
  .start("abfss://bronze@projectdatalakestrgacc.dfs.core.windows.net/netflix_titles")
```
